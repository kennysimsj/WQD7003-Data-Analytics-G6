Attempting to solve the business issue of:<br>

What are the indicators of depression, anxiety or stress among the respondents?<br>

By using sample test:

Importing dataset:

```python
import pandas as pd
from scipy.stats import ttest_rel
```

```python
# Read file using pandas
df = pd.read_csv('modified_dataset.csv')
df
```

```python
data = df[["depression_scores","anxiety_scores","stress_scores","depression_status","anxiety_status","stress_status"]]
print(data)
```

```python
#Dependent sample t-test
#the means of two related groups to determine whether there is a statistically significant difference between these means.
#comparing depression scores and anxiety scores 
a = df['depression_scores']
b = df['anxiety_scores']
t, p = ttest_rel(a, b)
print('t-statistic: ', t)
print('p-value: ', p)
print('Since the p-value is less than 0.05, reject null hypothesis')
print('Conclusion: There is a significant different in mean between depression scores and anxiety scores.')
```

```python
#comparing depression scores and stress scores 

a = df['depression_scores']
b = df['stress_scores']
t, p = ttest_rel(a, b)
print('t-statistic: ', t)
print('p-value: ', p)
print('Since the p-value is less than 0.05, reject null hypothesis')
print('Conclusion: There is a significant different in mean between depression scores and stress scores.')
```

```python
#comparing anxiety scores and stress scores 

a = df['anxiety_scores']
b = df['stress_scores']
t, p = ttest_rel(a, b)
print('t-statistic: ', t)
print('p-value: ', p)
print('Since the p-value is less than 0.05, reject null hypothesis')
print('Conclusion: There is a significant different in mean between anxiety scores and stress scores.')
```
